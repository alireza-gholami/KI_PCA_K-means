```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df_segmentation = pd.read_csv('p002_1.csv', sep=';', decimal=",")
df_segmentation.head()
```

```python
plt.figure(figsize = (12, 9))
plt.scatter(df_segmentation.iloc[:, 2], df_segmentation.iloc[:, 4])
plt.xlabel('Bundesland')
plt.ylabel('Physik')
plt.title('Visualisatin')
```

```python
scaler = StandardScaler()
num_columns = 16
# segmentation_std = scaler.fit_transform(df_segmentation[0:num_columns])
segmentation_std = scaler.fit_transform(df_segmentation)
```

```python
pca = PCA()
pca.fit(segmentation_std)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize = (10, 8))
# plt.plot(range(0,num_columns), pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')
plt.plot(range(0,16), pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')
plt.title('Explained Variance by Components')
plt.xlabel('number of Components')
plt.ylabel('Cumulative Explained Variance')
```

```python
pca = PCA(n_components = 3)
```

```python
pca.fit(segmentation_std)
```

```python
pca.transform(segmentation_std)
```

```python
scores_pca = pca.transform(segmentation_std)
```

```python
wcss = []
for i in range(1,16):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1,16), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means with PCA Clustering')
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
```

```python
kmeans_pca.fit(scores_pca)
```

```python
df_segm_pca_kmeans = pd.concat([df_segmentation.reset_index(drop = True), pd.DataFrame(scores_pca)], axis = 1)
# df_segm_pca_kmeans.columns.values[-5: ] = ['Component 1', 'Component 2', 'Component 3', 'Component 4', 'Component 5']
df_segm_pca_kmeans.columns.values[-3: ] = ['Component 1', 'Component 2', 'Component 3']
df_segm_pca_kmeans['Segment K-means PCA'] = kmeans_pca.labels_
```

```python
df_segm_pca_kmeans.head()
```

```python
df_segm_pca_kmeans['Segment'] = df_segm_pca_kmeans['Segment K-means PCA'].map({0:'first',
                                                                               1:'second',
                                                                               2:'third',
                                                                               3:'fourth'})
#                                                                                4:'fifth',
#                                                                                5:'sixth',
#                                                                                6:'seventh'})
# df_segm_pca_kmeans['Segment'] = df_segm_pca_kmeans['Segment K-means PCA'].map({0:'first',
#                                                                                1:'second'})
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import ListedColormap
x_axis = df_segm_pca_kmeans['Component 1']
y_axis = df_segm_pca_kmeans['Component 2']
z_axis = df_segm_pca_kmeans['Component 3']
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
cmap = ListedColormap(sns.color_palette("husl", 256).as_hex())
ax.scatter(x_axis, y_axis, z_axis, cmap="brg")
# sns.scatterplot(x_axis, y_axis, z_axis, hue = df_segm_pca_kmeans['Segment'], palette = ['r', 'g', 'b'])
plt.title('Cluster by PCA Components')
plt.show()
```
